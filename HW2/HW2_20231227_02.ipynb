```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

df = pd.read_csv('Boston_house.csv')
df
```

```python
selected_columns = ['B', 'RM', 'DIS','CHAS']

# 데이터프레임에서 선택한 열 추출
selected_data = df[selected_columns]

# 타겟 변수
y = df['Target']
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical

X_train, X_test, y_train, y_test = train_test_split(selected_data, y, test_size=0.2, random_state=42)

print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error, mean_absolute_error

rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)

y_pred = rfr.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)  # RMSE는 squared=False를 설정하여 얻을 수 있습니다.
mae = mean_absolute_error(y_test, y_pred)

print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'Mean Absolute Error (MAE): {mae}')

# Mean Squared Error (MSE): 26.204056382352906
# Root Mean Squared Error (RMSE): 5.1189897814268885
# Mean Absolute Error (MAE): 3.371107843137254
```

```python
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)

y_pred = dtr.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)  # RMSE는 squared=False를 설정하여 얻을 수 있습니다.
mae = mean_absolute_error(y_test, y_pred)

print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'Mean Absolute Error (MAE): {mae}')


# Mean Squared Error (MSE): 40.39843137254902
# Root Mean Squared Error (RMSE): 6.355976036184295
# Mean Absolute Error (MAE): 4.386274509803922
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)  # RMSE는 squared=False를 설정하여 얻을 수 있습니다.
mae = mean_absolute_error(y_test, y_pred)

print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'Mean Absolute Error (MAE): {mae}')

# Mean Squared Error (MSE): 42.93105924068644
# Root Mean Squared Error (RMSE): 6.5521797320194475
# Mean Absolute Error (MAE): 4.103651805693235
```
