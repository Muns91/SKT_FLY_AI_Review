# 인덱싱과 슬라이싱

인덱싱(Indexing) \
1. 데이터 프레임의 특정 위치의 요소에 접근한다.

슬라이싱(Slicing) \
1. 특정 부분의 데이터를을 추출하는 것
2. loc와 iloc를 사용
3. loc : loc[행 인덱싱, 열 인덱싱]의 형태로 주로 사용하지만, loc[행 인덱싱]의 형태로도 사용, 열은 컬럼명을 사용
4. iloc : 데이터가 있는 위치(정수)를 지정

```python
import numpy as np
import pandas as pd
z = [['Jackson', 68, True, np.NaN, 'A'],
['Liam', 74, True, 1.1, 'A'],
['Emma', np.NaN, False, 0.89, 'C'],
['James', 78, np.NaN, 0.91, 'C'],
['Ava', 54, False, 0.73, 'C'],
['Henry', 69, True, np.NaN, 'C']]
df = pd.DataFrame(data= z, columns=['Name', 'Weight', 'Option', 'Rate', 'Type'])
df
```

```python
df.loc[2:4]
```

```python
df.loc[2:4, ['Name', 'Weight'] ]
```

```python
df.loc[2:4, 'Weight']
```

```python
x = df.loc[df['Rate']>0.8, ['Weight', 'Rate', 'Type']]
print('type : ', type(x))
x
```

```python
x = df.loc[df['Option']==True]
x
```

```python
 x = df.loc[(df['Rate']>0.8) &(df['Option']==True), :]
 x
```

```python
x = df.iloc[2:4, 1:-1]
x
```

```python
x = [['Jackson', 68, True, 0.82, 'A'], ['Liam', 74, True, 1.1, 'B'], ['Emma', 52, False, 0.89, 'C'],
 ['James', 78, True, 0.91, 'C'], ['Ava', 54, False, 0.73, 'B'], ['Henry', 69, True, 0.78, 'C']]
df = pd.DataFrame(x, columns=['Name', 'Weight', 'Option', 'Rate', 'Type'])
df.info()
```

```python
x = df.select_dtypes(include=['number'])
x
```

# 데이터 프레임 변경

구조 변경 \
1. 컬럼을 추가/삭제
2. 컬럼명 변경
3. 특정 부분의 값 변경
4. 인덱스 변경

```python
x = [['Jackson', 68, True],
['Liam', 74, True],
['Emma', 52, False],
['James', 78, True],
['Ava', 54, False],
['Henry', 69, True]]
df = pd.DataFrame(x, columns=['Name', 'Weight', 'Option'])
df
```

```python
# 컬럼 추가
df['Type'] = 'A'
df
```

```python
# 컬럼명 변경
df.rename(columns={'Type': 'Grade'}, inplace=True)
df
```

```python
# 컬럼명 전체 변경
df.columns = ['A', 'B', 'C', 'D']
df
```

```python
x = [['Jackson', 68, True, 0.82, 'A'],
['Liam', 74, True, 1.1, 'B'],
['Emma', 52, False, 0.89, 'C'],
['James', 78, True, 0.91, 'C'],
['Ava', 54, False, 0.73, 'B'],
['Henry', 69, True, 0.78, 'C']]
df = pd.DataFrame(x, columns=['Name', 'Weight', 'Option', 'Rate', 'Type'])
df
```

```python
# 특정 값 변경
df.loc[1, 'Weight'] = 100
df
```

```python
# 특정 부분 값 수정
df.loc[df['Option']==False, 'Type'] = np.nan
df
```

```python
# 컬럼 삭제
df.drop(['Option', 'Weight'], axis=1, inplace=True)
df
```

```python
df = pd.DataFrame({
'Weight':['68', '74','52','78','54','69'],
'Option':[True, True, False, True, False, True],
'Rate':[0.82, 1.1, 0.89, 0.91, 0.73, 0.78],
'Type':['A','B','C','C','B','C']
})
df.info()
```

```python
# to_numeric() : ’Weight’ 컬럼을 문자열(object)에서 숫자로 변경

df['Weight'] = pd.to_numeric(df['Weight'])
df.info()
```

```python
#  astype() : ’Weight' 컬럼을 숫자에서 문자열로 변경
df['Weight'] = df['Weight'].astype(str)
df.info()
```

# 그룹핑 함수

```python
x = df.groupby('city')['price'].mean()
x
```

```python
x = df.groupby(['city', 'fruits']).mean()
x
```
